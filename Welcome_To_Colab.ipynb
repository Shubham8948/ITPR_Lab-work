```
 # Input number
num1 = int(input("Enter first number :"))
num2 = int(input("Enter second number :"))

#-----------------------------------
if(num1 == 0):
    print ("HCF: ",num1)
elif(num2 == 0):
    print ("HCF: ",num2)
else:
    while(num2 != 0):
        rem = num1 % num2
        num1 = num2
        num2 = rem
    #----------------------------------
    print("HCF :", num1)
```

```
# Input number
num = int(input("Enter a number: "))
num = abs(num)
max_digit = 0
while num > 0:
    digit = num % 10
    if digit > max_digit:
        max_digit = digit
    num = num // 10

print("Greatest digit:", max_digit)
```
